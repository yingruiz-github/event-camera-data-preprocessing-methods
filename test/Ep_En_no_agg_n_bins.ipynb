#### get file path
RAW_FILE_PATH: path to the raw event data file
RAW_META_PATH: path to the ground truth metadata file. This file has two columns seperated by space. column1 is timestamp the
               ground truth image being taken (normal image), the second columns is the image path
POS_DIR: dir used to save proccessed positive img
NEG_DIR: dir used to save processed negative img
EXAMPLE_IMG_PATH: an example of ground truth image

```python
RAW_FILE_PATH = "../data/raw_events/events.txt"
RAW_META_PATH = "../data/raw_events/images.txt"
POS_DIR = "../data/processed_imgs/pos"
NEG_DIR = "../data/processed_imgs/neg"
EXAMPLE_IMG_PATH = "../data/example_img/frame_00000000.png"
```

####  read the ground truth time stamp

```python
ground_truth_timestamps = []
with open(RAW_META_PATH, "r") as f:
    lines = f.readlines()
    for line in lines:
        ground_truth_timestamps.append(line.split()[0])
```

```python
ground_truth_timestamps = [float(i) for i in ground_truth_timestamps]
ground_truth_timestamps
```

#### extract raw event data to a pandas df

```python
df = extract_data(os.path.normpath(RAW_FILE_PATH))
```
